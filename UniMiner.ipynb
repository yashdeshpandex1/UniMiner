```python
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from webdriver_manager.firefox import GeckoDriverManager
from selenium.webdriver.common.by import By 
import csv
import time
```

```python
url = "https://www.unirank.org/fr/a-z/"
options = webdriver.FirefoxOptions()
options.add_argument("--headless")
```

```python
driver = webdriver.Firefox(service=Service(GeckoDriverManager().install()), options=options)
driver.get(url)
time.sleep(4)
```

```python
rows = driver.find_elements(By.CSS_SELECTOR, "table tbody tr")

universities = []

for row in rows:
    cols = row.find_elements(By.TAG_NAME, "td")
    if len(cols) >= 3:
        rank  = cols[0].text.strip()
        name  = cols[1].text.strip()
        town  = cols[2].text.strip()
        universities.append([rank, name, town])

driver.quit()
```

```python
with open("unirank_france_universities.csv", "w", newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(["rank", "name", "town"])
    writer.writerows(universities)

print(f"{len(universities)} universities saved.")
```
